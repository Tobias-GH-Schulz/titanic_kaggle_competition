```python
%timeit

from pycaret.classification import *
import pandas as pd
import numpy as np
from catboost import Pool, CatBoostClassifier
from xgboost import XGBClassifier

data = pd.read_csv('Collection/titanic_comp/Train_clean.csv')

grid = setup(data, 
             target='Survived', 
             log_experiment=True, 
            
             silent=True, 
             verbose=True,
             log_plots=True,
             n_jobs=1)
```
