```
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neural_network import MLPClassifier

#from xgboost import XGBClassifier
from sklearn import model_selection
from sklearn.utils import class_weight

import numpy as np
import pandas as pd



train = pd.read_csv('Train_clean.csv', index_col=0)
test = pd.read_csv('Test_clean.csv', index_col=0)
train  = train.drop("not_alone", axis=1)
test  = test.drop("not_alone", axis=1)

train = train.drop("Parch", axis=1)
test  = test.drop("Parch", axis=1)
```

```
#test['Fare'].fillna(0, inplace=True)
```

```
X_train = train.drop(columns=['Survived'], axis=1)
y_train = train.Survived
```

```
scaler = StandardScaler()
scaler.fit(X_train)
X_train_norm = scaler.transform(X_train)
test_norm = scaler.transform(test)
```

```
rf = RandomForestClassifier(criterion = "gini", 
                                       min_samples_leaf = 1, 
                                       min_samples_split = 10,   
                                       n_estimators=100, 
                                       max_features='auto', 
                                       oob_score=True, 
                                       random_state=1, 
                                       n_jobs=-1)
rf.fit(X_train_norm,y_train)

rf_pred = rf.predict(test_norm)
rf_pred
```

```
knn = KNeighborsClassifier()
knn.fit(X_train_norm,y_train)
knn_pred = knn.predict(test_norm)
```

```
svclassifier = SVC(kernel='poly')
sv = svclassifier.fit(X_train_norm, y_train)
# print(sv.score(X_train, y_train))
svc_pred = svclassifier.predict(test_norm)
svc_pred
```

```
rf = RandomForestClassifier()
rf.fit(X_train_norm,y_train)

rf2_pred = rf.predict(test_norm)
rf2_pred
```

```
clf = MLPClassifier(activation= 'tanh', alpha= 0.01, random_state= 0)
clf.fit(X_train_norm, y_train)
```

```
mlp_pred = clf.predict(test_norm)
```
